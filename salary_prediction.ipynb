```python
#import Libraries
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
df = pd.read_csv('salaries.csv')  #Read The Csv File Here
df
```

```python
y = df.salary_more_then_100k                    #Assign a variable to the lables of dataset
x = df.drop('salary_more_then_100k', axis = 1)  #Drop the labels for prediction
x
```

```python
#Use the Label Encoder To Convert The Data From String To Int
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
x['le_company'] = le_company.fit_transform(x['company'])
x['le_job'] = le_job.fit_transform(x['job'])
x['le_degre'] = le_degree.fit_transform(x['degree'])
```

```python
x
```

```python
x = x.drop (['company','job','degree'], axis = 1)
x
```

```python
y.values
```

```python
model = DecisionTreeClassifier()            #Apply The Model on The Features
model.fit(x.values, y)
```

```python
model.score(x.values,y)
```

```python
p1 = model.predict([[2,1,0]])
print(str(p1)+ ' = No')
```

```python
p2 = model.predict([[1,1,1]])
print(str(p1)+ ' = Yes')
```
